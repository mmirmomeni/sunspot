```python
import research as r
import scipy
import pandas

def quick_look(path, treatment):
    W = r.load_files(r.find_files(path, treatment+"._.*fitness.dat"))
    W['rmse'] = 100.0 / W['max_fitness'] - 1.0

    figure()
    for i,g in W.groupby('trial'):
        plot(g['update'], g['rmse'])
    title('sum rmse, t=1:8')
    
    final = W[W['update']==W['update'].max()]
    print "Dominant RMSE:"
    print final.ix[final['rmse'].idxmin()]
    
```
